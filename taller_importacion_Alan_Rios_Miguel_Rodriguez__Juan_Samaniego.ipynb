```python
indicador = {"FP.CPI.TOTL.ZG": "Inflación"}

fecha_inicio = datetime.datetime(2010,1,1)
fecha_fin = datetime.datetime(2023,1,1)

paises = ["COL","ECU","BRA","CHL","MEX","PER"]

inflacion = wbdata.get_dataframe(indicador, country=paises, date=(fecha_inicio, fecha_fin))
inflacion = inflacion.reset_index()


print(inflacion.head())
```
