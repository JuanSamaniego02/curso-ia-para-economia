## **Ejercicio 4**

Un cajero autom√°tico tiene 500.000 pesos disponibles.
Un usuario puede retirar varias veces, pero no puede retirar m√°s de lo que queda en el cajero. Escribe un programa que:

- Pregunte cu√°nto quiere retirar.

- Reste ese valor al saldo.

- Termine cuando el saldo llegue a 0 o el usuario escriba 0.

```python
saldo = 500000

print("Saldo disponible:", saldo)

while saldo > 0:
    retiro = int(input("¬øCu√°nto dinero desea retirar? (Escriba 0 para salir): "))

    if retiro == 0:
        print("Operaci√≥n finalizada por el usuario.")
        break

    if retiro > saldo:
        print("No puede retirar m√°s dinero del disponible.")
        print("Saldo actual:", saldo)
    elif retiro < 0:
        print("Ingrese un valor v√°lido.")
    else:
        saldo -= retiro
        print("Retiro exitoso.")
        print("Saldo restante:", saldo)

if saldo == 0:
    print("El cajero se qued√≥ sin dinero.")
```

## **Ejercicio 7**

Escribe un c√≥digo que permita al usuario crear una contrase√±a y que verifique que esta cumpla todas las siguientes condiciones:

- Tiene al menos 8 caracteres.

- Contiene al menos una letra may√∫scula.

- Contiene al menos una letra min√∫scula.

- Contiene al menos un n√∫mero.

Si cumple todas las condiciones, muestra "Contrase√±a v√°lida", de lo contrario, explica qu√© condici√≥n no se cumple.

```python
contrase√±a = input("Crea tu contrase√±a: ")

errores = []

if len(contrase√±a) < 8:
    errores.append("Debe tener m√≠nimo 8 caracteres")

if not any(c.isupper() for c in contrase√±a):
    errores.append("Debe tener al menos una may√∫scula")

if not any(c.islower() for c in contrase√±a):
    errores.append("Debe tener al menos una min√∫scula")

if not any(c.isdigit() for c in contrase√±a):
    errores.append("Debe tener al menos un n√∫mero")

if len(errores) == 0:
    print("Contrase√±a v√°lida ‚úÖ")
else:
    for error in errores:
      print(error)
```

Crea una funci√≥n llamada "extraer_dominio" que extraiga el dominio de una p√°gina web a partir de una cadena que contenga un email de la siguiente forma:

```python
def extraer_dominio(dominio):
    """
    Extrae el dominio de una direcci√≥n de correo electr√≥nico.

    Par√°metros:
    email (str): Cadena de texto que representa un email
                 en el formato usuario@dominio.com

    Retorna:
    str:
        - El dominio del email si el formato es v√°lido.
        - "Email inv√°lido" si no contiene el s√≠mbolo @.
    """

    if "@" in dominio:
        return dominio.split("@")[1]
    else:
        return "Email inv√°lido"
```

```python
def contar_desempleo(frase):
    """
    Cuenta cu√°ntas veces aparece la palabra "desempleo" en un texto,
    sin importar si est√° en may√∫sculas o min√∫sculas.

    Par√°metros:
    frase (str): Cadena de texto en la que se desea buscar la palabra.

    Retorna:
    int: N√∫mero de veces que aparece la palabra "desempleo" en el texto.
    """
    frase = frase.lower()
    veces = frase.count("desempleo")
    return veces
```

## **Ejercicio 10**

Vas conduciendo un carro muy r√°pido y eres detenido por un agente de tr√°nsito.

Escribe una funci√≥n "multa" que retorne uno de los 3 siguientes resultados:  

- Si la velocidad es 60 Km/h o menos, el resultado es "Sin multa".
- Si la velocidad est√° entre 60 y 80 Km/h (inclusive), el resultado es "Multa peque√±a".
- Si la velocidad es mayor a 80 Km/h, el resultado es "Multa grande".

Adem√°s, considera que si esto ocurre el d√≠a de tu cumplea√±os (codificado como par√°metro booleano que recibe la funci√≥n), tu velocidad podr√≠a ser 5 Km/ mayor en todos los casos.

```python
def multa(velocidad,cumpleanos):
    """
    Determina el tipo de multa seg√∫n la velocidad del veh√≠culo.

    Par√°metros:
    velocidad (int o float): Velocidad a la que se conduce (en Km/h).
    cumpleanos (bool): Indica si es el cumplea√±os del conductor.
                       Si es True, se permiten 5 Km/h adicionales.

    Retorna:
    str:
        - "Sin multa" si la velocidad es 60 Km/h o menos.
        - "Multa peque√±a" si la velocidad est√° entre 61 y 80 Km/h (inclusive).
        - "Multa grande" si la velocidad es mayor a 80 Km/h.
    """
    if cumpleanos == True:
      velocidad += 5
    if velocidad <= 60 :
      print("Sin multa")
    if 60< velocidad <= 80 :
      print("Multa Peque√±a")
    if velocidad > 80 :
      print("Multa grande")
```

```python
multa(81, True)
```

```python
multa(81, False)
```

```python
multa(59, False)
```

## **Ejercicio 14**

Tienes un diccionario con exportaciones en millones de USD. Escribe un ciclo for que recorra los valores, calcule el promedio de exportaciones y lo imprima como: "El promedio de exportaciones es: XXXX millones de USD"

```python
exportaciones = {"Colombia": 1500, "M√©xico": 2300, "Chile": 1200, "Per√∫": 1800}
```

```python
suma = 0
contador = 0

for valor in exportaciones.values():
    suma += valor
    contador += 1

promedio = suma / contador

print(f"El promedio de exportaciones es: {promedio} millones de USD")
```

## **Ejercicio 16**

Una empresa quiere vender al menos 100 unidades de un producto. Cada d√≠a vende un n√∫mero aleatorio de unidades entre 1 y 10 (pista: usa import random y random.randint(1,10)). Escribe un programa con while que muestre:

- El total vendido cada d√≠a.

- En cu√°ntos d√≠as se logr√≥ la meta.

```python
import random
total_vendido = 0
dias = 0
while total_vendido <= 100:
  total_vendido += random.randint(1,10)
  dias += 1



print(f"\nMeta alcanzada en {dias} d√≠as con un total de {total_vendido} unidades vendidas.")
```
